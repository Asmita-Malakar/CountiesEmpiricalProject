```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
counties_data = pd.read_csv('counties.csv')
```

```python
# Political difference between different regions
#4 Southern States
Southern_Regions1 = counties_data.loc[counties_data["state"] == "MD", "democrat"].median()
Southern_Regions2 = counties_data.loc[counties_data["state"] == "TX", "democrat"].median()
Southern_Regions3 = counties_data.loc[counties_data["state"] == "AL", "democrat"].median()
Southern_Regions4 = counties_data.loc[counties_data["state"] == "SC", "democrat"].median()
print(Southern_Regions1)
print(Southern_Regions2)
print(Southern_Regions3)
print(Southern_Regions4)
Northern_Regions1 = counties_data.loc[counties_data["state"] == "MA", "democrat"].median()
Northern_Regions2 = counties_data.loc[counties_data["state"] == "CA", "democrat"].median()
Northern_Regions3 = counties_data.loc[counties_data["state"] == "NY", "democrat"].median()
Northern_Regions4 = counties_data.loc[counties_data["state"] == "VT", "democrat"].median()
print(Northern_Regions1)
print(Northern_Regions2)
print(Northern_Regions3)
print(Northern_Regions4)
```

```python
counties_data["college_grad_fraction"] = counties_data["college"]/100
```

```python
counties_data["college_grad_fraction"].head()
```

```python
X = counties_data["college_grad_fraction"]
y = counties_data["income"]
df = pd.DataFrame(
    {'X': X,
     'y': y}
)
df.head()
plt.figure(figsize=(12, 6))
plt.plot(X, y, 'ro')   # scatter plot showing actual data
plt.title('College Graduation impact on income')
plt.xlabel('College Graduation')
plt.ylabel('Income')

plt.show()
#Positive Correlation
```

```python
X = counties_data["white"]
y = counties_data["college_grad_fraction"]
df = pd.DataFrame(
    {'X': X,
     'y': y}
)
df.head()

plt.figure(figsize=(12, 6))
plt.plot(X, counties_data["college_grad_fraction"]) 
plt.plot(X, y, 'ro')   # scatter plot showing actual data
plt.title('Race impact on college Graduation')
plt.xlabel('White population percentage')
plt.ylabel('College Grad rate')

plt.show()
#Moderate Correlation
X = counties_data["black"]
y = counties_data["college_grad_fraction"]
df = pd.DataFrame(
    {'X': X,
     'y': y}
)
df.head()

plt.figure(figsize=(12, 6))
plt.plot(X, counties_data["college_grad_fraction"]) 
plt.plot(X, y, 'ro')   # scatter plot showing actual data
plt.title('Race impact on college Graduation')
plt.xlabel('Black population percentage')
plt.ylabel('College Grad rate')

plt.show()


#Importance of Race in college graduation
```
